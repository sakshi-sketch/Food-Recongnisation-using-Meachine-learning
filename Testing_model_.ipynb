Loading model

```python
cnn = tf.keras.models.load_model('/content/trained_model.h5')
```

performing single prediction

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/food recongnization system/test',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False,

)
```

```python
import cv2 as cv
```

```python
path = '/content/drive/MyDrive/food recongnization system/test/grapes/Image_2.jpg'
img = cv.imread(path,1)
plt.imshow(img)
plt.title('test image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
path1 = '/content/drive/MyDrive/food recongnization system/test/banana/Image_10.jpg'
img1 = cv.imread(path1,1)
plt.imshow(img1)
plt.title('test image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image = tf.keras.preprocessing.image.load_img(path,target_size = (64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr]) # [ ] = for converting single image in 2 dimensional array
predictions = cnn.predict(input_arr)
```

```python
image1 = tf.keras.preprocessing.image.load_img(path1,target_size = (64,64))
input_arr1 = tf.keras.preprocessing.image.img_to_array(image1)
input_arr1 = np.array([input_arr1]) # [ ] = for converting single image in 2 dimensional array
predictions1 = cnn.predict(input_arr1)
```
